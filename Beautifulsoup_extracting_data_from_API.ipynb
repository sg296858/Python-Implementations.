```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r=requests.get("https://www.iplt20.com/auction/2022")
r
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
table=soup.find("table",class_="ih-td-tab auction-tbl")
print(table)
```

```python
title=table.find_all("th")
```

```python
header=[]
for i in title:
    name=i.text
    header.append(name)
```

```python
header
```

```python
df=pd.DataFrame(columns=header)
df
```

```python
rows =table.find_all('tr')
```

```python
for i in rows[1:]:
  data=i.find_all('td')
  row=[tr.text for tr in data]
  l=len(df)
  df.loc[l]=row
```

```python
df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
response=requests.get("https://pokeapi.co/api/v2/")
```

```python
response
```

```python
api_dict=response.json()
```

```python
def fetch_data(api_url):
    response = requests.get(api_url)
    if response.status_code == 200:
        return response.json()  # Return JSON response if status is OK
    else:
        return None  # Return None if the API request fails

# List to store the data for each category
data_list = []

# Iterate through the dictionary of API endpoints and fetch data
for category, url in api_dict.items():
    data = fetch_data(url)
    if data:
        # Here, we add the category as a key to associate the data with its name
        data['category'] = category
        data_list.append(data)

# Convert list of dictionaries to DataFrame
df = pd.DataFrame(data_list)

# Show the DataFrame (can use .head() to limit the output)
print(df.head())
```
